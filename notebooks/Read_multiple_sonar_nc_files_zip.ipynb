```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_mfdataset('sample_Sv/*_Sv.nc')
ds
```

`ds` is an `xarray.Dataset` which has 3 dimensions: `frequency`, `ping_time` and `range_bin`.

To plot an "echogram" -- image formed by echoes from multiple sonar pings -- we can select one frequency (for example 18 kHz below), and display the `Sv` as a 2 dimensional image with `ping_time` in the horizontal axis and range_bin (which is related to ocean depth) in the vertical axis.

Note:
- `Sv` represents volume backscatter strength. Its distribution in time and space is what researchers are interested in.
- There is another data variable `range` in this dataset, which is the distance of echoes counting from the face of the trasducer, in meters. For the visualization to be physically meaningful, the vertical axis should be changed to use `range`.

```python
plt.figure(figsize = (20, 5))
ds.Sv.sel(frequency=18000).T.plot(yincrease=False, vmin=-80, vmax=-40, cmap='viridis')
```
